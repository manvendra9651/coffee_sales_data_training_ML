```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df =  pd.read_csv(r"C:\\Users\DELL\Downloads\archive (3)\Sales_coffee.csv")
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['card'].fillna("unknown" , inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['hour'] = df['datetime'].dt.hour
```

```python
df.head()
```

```python
le = LabelEncoder()
df['cash_type'] = le.fit_transform(df['cash_type'])
df['coffee_name'] = le.fit_transform(df['coffee_name'])
```

```python
x = df[['hour','cash_type','money']]
y = df['coffee_name']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```
